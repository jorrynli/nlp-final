```python
import matplotlib.pyplot as plt
```

```python

dropout_rates = [0, 0.1, 0.2, 0.25, 0.3, 0.35, 0.5]        # x-axis: dropout rates
val_perplexities = [112.9, 91.4, 86.7, 86.0, 84.5, 84.9, 86.1]  # y-axis: validation perplexities

# ----------------------------------------------------

plt.figure(figsize=(8, 5))
plt.plot(dropout_rates, val_perplexities, marker='o', linestyle='-', linewidth=2)

plt.xlabel("Dropout Rate", fontsize=12)
plt.ylabel("Validation Perplexity", fontsize=12)
plt.title("Dropout Rate vs Validation Perplexity, ByteLevelBPETokenizer", fontsize=14)

plt.grid(True, linestyle="--", alpha=0.6)
plt.tight_layout()
plt.show()
```
